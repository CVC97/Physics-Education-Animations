```python
%%capture_video --path "animations/parametrization/parametrization_ENG_C.mp4"
%%manim -qh --fps 60 $video_scene

class para_Scene(ThreeDScene):
    def construct(self):
        CVC = Text('CVC', font_size = 12, weight = BOLD, color = WHITE, font = 'Latin Modern Sans').align_on_border(RIGHT + DOWN, buff = 0.2)
        self.add(CVC)

         # Headline
        text_parametrization = Tex(r"Parametrization of a Trajectory in $\mathbb{R}^3$", font_size = 48).align_on_border(UP + LEFT, buff = 0.5).shift(0.5 * RIGHT)

        text_para_def = MathTex(r"t\in\left[t_0,\,t_1\right]", font_size = 48).move_to([0.5, 0, 0])
        text_para_def_int = MathTex(r"t\in\left[0,\,10\pi\right]", font_size = 48).move_to([0.5, 0, 0])
        text_para_left = MathTex(r",\quad t\mapsto", font_size = 48).next_to(text_para_def_int, RIGHT)
        text_para_right = Matrix([[r'x(t)'], [r'y(t)'], [r'z(t)']]).scale(0.85).next_to(text_para_left, RIGHT)
        text_para_right_end = Matrix([[r'2\cos{(t)}'], [r'\sin{(t)}'], [r't/(2\pi)']]).scale(0.85).next_to(text_para_left, RIGHT)
        text_para_right_end_white = Matrix([[r'2\cos{(t)}'], [r'\sin{(t)}'], [r't/(2\pi)']]).scale(0.85).next_to(text_para_left, RIGHT)
        text_para_right[0][0].set_color(BLUE)
        text_para_right[0][1].set_color(YELLOW)
        text_para_right[0][2].set_color(RED)
        text_para_right_end[0][0].set_color(BLUE)
        text_para_right_end[0][1].set_color(YELLOW)
        text_para_right_end[0][2].set_color(RED)
        mapping_group = VGroup(text_para_left, text_para_right, text_para_def)
        mapping_group_int = VGroup(text_para_left, text_para_right, text_para_def_int)
        mapping_group_end = VGroup(text_para_left, text_para_right_end, text_para_def_int)
        mapping_group_end_white = VGroup(text_para_left, text_para_right_end_white, text_para_def_int)


        # 3D-Coordinate System
        CO3D = [-6, 0, -8]
        CO3D_x_range = (-3, 3, 1)
        CO3D_y_range = (-3, 3, 1)
        CO3D_z_range = (-1, 6, 1)
        ax = ThreeDAxes(
            x_range = CO3D_x_range, y_range = CO3D_y_range, z_range = CO3D_z_range,
            x_length = 6, y_length = 6, z_length = 9, axis_config = {'tip_length': 0.05, 'tip_width': 0.3}, 
            z_axis_config = {'color': WHITE},
            ).move_to(CO3D).rotate(
                axis = [1, 0, 0], angle = 6*PI/4
                ).rotate(
                    axis = [0, 1, 0], angle = 5*PI/4
                    ).rotate(axis = [1, 0, 0], angle = PI/12).rotate(
                        axis = [0, 1, 0], angle = PI/16
                        )
        x_lab = Tex('$x$', font_size = 64).move_to(CO3D).shift(4*DOWN + 2.75*LEFT)
        y_lab = Tex('$y$', font_size = 64).move_to(CO3D).shift(4*DOWN + 2.75*RIGHT)
        z_lab = Tex('$z$', font_size = 64).move_to(CO3D + 0.5*RIGHT).shift(4.8*UP)
        ax_3D = VGroup(ax, x_lab, y_lab, z_lab).set_opacity(0.5)


        # components of the parametric function
        def x_func(t):
            return 2 * np.cos(t)

        def y_func(t):
            return 1 * np.sin(t)
            
        def z_func(t):
            return t / 2 / PI


        # parametric function specified to ax
        def para_ax_func(t):
            return ax.c2p(x_func(t), y_func(t), z_func(t))

        # generates t_line
        def t_line(t, color):
            return Line3D(para_ax_func(t) + np.array([0, 0.25, 0]), para_ax_func(t) + np.array([0, -0.25, 0]), color = color)

        # generates connector
        def r_connector(start, end, color = RED):
            return Line(start, end, color = color, stroke_width = 4)


        # function start and end
        t0 = 0
        t1 = 10 * PI
        t_bar = 6.75 * PI
        para_curve = ParametricFunction(para_ax_func, color = GREY, t_range = np.array([t0, t1, 0.01])).set_shade_in_3d(True)
        t0_line = t_line(t0, WHITE)
        t1_line = t_line(t1, WHITE)
        t0_label = MathTex(r't_0', font_size = 36, stroke_width = 1, color = WHITE).next_to(t0_line, DOWN)
        t1_label = MathTex(r't_1', font_size = 36, stroke_width = 1, color = WHITE).next_to(t1_line, UP)


        # 3D sphere and connectors
        r_sphere = Sphere(center = para_ax_func(0), radius = 0.1, resolution = (10, 10))
        r_sphere.set_color(GREY)
        #r_sphere.z_index = 5
        x = x_func(0)
        y = y_func(0)
        z = z_func(0)
        x_line = r_connector(ax.c2p(0, y, 0), ax.c2p(x, y, 0), color = BLUE)
        x_line.z_index = 0
        y_line = r_connector(ax.c2p(x, 0, 0), ax.c2p(x, y, 0), color = YELLOW)
        z_line = r_connector(ax.c2p(x, y, z), ax.c2p(x, y, 0), color = RED)
        x_label = MathTex(r"x(t)", font_size = 36, color = BLUE).move_to([-2.75, -2.25, 0])
        y_label = MathTex(r"y(t)", font_size = 36, color = YELLOW).move_to([-3.75, -1.65, 0])
        z_label = MathTex(r"z(t)", font_size = 36, color = RED).move_to([-2.6, -0.5, 0])

        r_sphere_bar = Sphere(center = para_ax_func(t_bar), radius = 0.1, resolution = (10, 10))
        r_sphere_bar.set_color(GREY)
        #r_sphere_bar.z_index = 5
        x_bar = x_func(t_bar)
        y_bar = y_func(t_bar)
        z_bar = z_func(t_bar)
        x_line_bar = r_connector(ax.c2p(0, y_bar, 0), ax.c2p(x_bar, y_bar, 0), color = BLUE)
        x_line_bar.z_index = 0
        y_line_bar = r_connector(ax.c2p(x_bar, 0, 0), ax.c2p(x_bar, y_bar, 0), color = YELLOW)
        z_line_bar = r_connector(ax.c2p(x_bar, y_bar, z_bar), ax.c2p(x_bar, y_bar, 0), color = RED)
        t_bar_group = VGroup(r_sphere_bar, x_line_bar, y_line_bar, z_line_bar)
        t_bar_labels = VGroup(x_label, y_label, z_label)


        # 2D-Coordinate Sytem for each coordinate
        origin_x = [2.5, 2.5, 0]
        origin_y = [2.5, 0, 0]
        origin_z = [2.5, -2.5, 0]
        x_length = 7
        y_length = 1.75
        x_range = [-0.5, 10.5, 1]
        x_dict = dict(zip([5, 10], [r"$5\pi$", r"$10\pi$"]))

        # x(t) system
        y_range_x = [-2.5, 2.5, 1]                          
        y_dict_x = dict(zip([-2, 0, 2], [-2, 0, 2]))
        npla_x = NumberPlane(
            x_range = x_range, y_range = y_range_x, x_length = x_length, y_length = y_length, 
            axis_config = {"include_numbers": False, "stroke_opacity": 0.25, "label_direction": DL}, y_axis_config = {"label_direction": DL}, 
            background_line_style = {"stroke_opacity": 0.25}, ).move_to(origin_x).add_coordinates(x_dict, y_dict_x)#.set_color(BLUE)
        npla_xlabel_x = npla_x.get_x_axis_label(Tex(r"$t$", font_size = 28)).shift(0.1 *LEFT)
        npla_ylabel_x = npla_x.get_y_axis_label(Tex(r"$x(t)$", font_size = 28, color = BLUE))#.shift(1.05 * LEFT + 0.15 * DOWN)
        npla_ylabel_x_ext = Tex(r"$=~2\cos{(t)}$", font_size = 28, color = BLUE).next_to(npla_ylabel_x, RIGHT)
        npla_x_group = VGroup(npla_x, npla_xlabel_x, npla_ylabel_x)

        # y(t) system
        y_range_y = [-1.25, 1.25, 1]                          
        y_dict_y = dict(zip([-1, 0, 1], [-1, 0, 1]))
        npla_y = NumberPlane(
            x_range = x_range, y_range = y_range_y, x_length = x_length, y_length = y_length, 
            axis_config = {"include_numbers": False, "stroke_opacity": 0.25, "label_direction": DL}, y_axis_config = {"label_direction": DL}, 
            background_line_style = {"stroke_opacity": 0.25}, ).move_to(origin_y).add_coordinates(x_dict, y_dict_y)
        npla_xlabel_y = npla_y.get_x_axis_label(Tex(r"$t$", font_size = 28)).shift(0.1 *LEFT)
        npla_ylabel_y = npla_y.get_y_axis_label(Tex(r"$y(t)$", font_size = 28, color = YELLOW))#.shift(1.05 * LEFT + 0.15 * DOWN)
        npla_ylabel_y_ext = Tex(r"$=~\sin{(t)}$", font_size = 28, color = YELLOW).next_to(npla_ylabel_y, RIGHT)
        npla_y_group = VGroup(npla_y, npla_xlabel_y, npla_ylabel_y)

        # z(t) system
        y_range_z = [-0.5, 5.5, 1]                          
        y_dict_z = dict(zip([0, 5], [0, 5]))
        npla_z = NumberPlane(
            x_range = x_range, y_range = y_range_z, x_length = x_length, y_length = y_length, 
            axis_config = {"include_numbers": False, "stroke_opacity": 0.25, "label_direction": DL}, y_axis_config = {"label_direction": DL}, 
            background_line_style = {"stroke_opacity": 0.25}, ).move_to(origin_z).add_coordinates(x_dict, y_dict_z)
        npla_xlabel_z = npla_z.get_x_axis_label(Tex(r"$t$", font_size = 28)).shift(0.1 *LEFT)
        npla_ylabel_z = npla_z.get_y_axis_label(Tex(r"$z(t)$", font_size = 28, color = RED))#.shift(0.9 * LEFT + 0.15 * DOWN)
        npla_ylabel_z_ext = Tex(r"$=~\frac{t}{2\pi}$", font_size = 28, color = RED).next_to(npla_ylabel_z, RIGHT)
        npla_z_group = VGroup(npla_z, npla_xlabel_z, npla_ylabel_z)


        # 2D-functions plus their updaters
        x_func_plot = npla_x.plot(lambda t_var: x_func(t_var * PI), color = BLUE, stroke_width = 1.5, stroke_opacity = 1, x_range = [0, 0])
        y_func_plot = npla_y.plot(lambda t_var: y_func(t_var * PI), color = YELLOW, stroke_width = 1.5, stroke_opacity = 1, x_range = [0, 0])
        z_func_plot = npla_z.plot(lambda t_var: z_func(t_var * PI), color = RED, stroke_width = 1.5, stroke_opacity = 1, x_range = [0, 0])

        def x_func_updater(func):
            t = t_tracker.get_value()
            func.become(npla_x.plot(lambda t_var: x_func(t_var * PI), color = BLUE, stroke_width = 1.5, stroke_opacity = 1, x_range = [0, 0.0001 + t / PI]))

        def y_func_updater(func):
            t = t_tracker.get_value()
            func.become(npla_y.plot(lambda t_var: y_func(t_var * PI), color = YELLOW, stroke_width = 1.5, stroke_opacity = 1, x_range = [0, 0.0001 + t / PI]))

        def z_func_updater(func):
            t = t_tracker.get_value()
            func.become(npla_z.plot(lambda t_var: z_func(t_var * PI), color = RED, stroke_width = 1.5, stroke_opacity = 1, x_range = [0, 0.0001 + t / PI]))


        # updates sphere and connectors
        def r_updater(sphere):
            t = t_tracker.get_value()
            x = x_func(t)
            y = y_func(t)
            z = z_func(t)
            sphere.move_to(para_ax_func(t))
       
        def x_line_updater(line):
            t = t_tracker.get_value()
            x = x_func(t)
            y = y_func(t)
            z = z_func(t)
            line.become(r_connector(ax.c2p(0, y, 0), ax.c2p(x, y, 0), color = BLUE))

        def y_line_updater(line):
            t = t_tracker.get_value()
            x = x_func(t)
            y = y_func(t)
            z = z_func(t)
            line.become(r_connector(ax.c2p(x, 0, 0), ax.c2p(x, y, 0), color = YELLOW))

        def z_line_updater(line):
            t = t_tracker.get_value()
            x = x_func(t)
            y = y_func(t)
            z = z_func(t)
            line.become(r_connector(ax.c2p(x, y, z), ax.c2p(x, y, 0), color = RED))


# Aligned Animation Part A
        def PartA():
            self.wait(1)
            self.play(Write(text_parametrization), run_time = 2)
            self.wait(1.5)
            self.play(FadeOut(text_parametrization), FadeIn(ax_3D), run_time = 3)
            self.play(Create(para_curve), run_time = 4)
            self.wait(4.5)
            self.play(Write(text_para_def), run_time = 1.5)
            self.wait(2.5)
            self.play(Write(t0_label), FadeIn(t0_line), run_time = 1.5)
            self.play(Write(t1_label), FadeIn(t1_line), run_time = 1.5)
            self.wait(9)
            self.play(FadeTransform(text_para_def, text_para_def_int), run_time = 3)
            self.wait(1.5)


# Aligned Animation Part B
        def PartB():
            self.add(ax_3D, para_curve, t0_label, t0_line, t1_label, t1_line, text_para_def_int)
            self.wait(2)
            self.play(FadeIn(t_bar_group), run_time = 5)
            self.play(FadeIn(t_bar_labels), run_time = 3)
            self.wait(4)
            self.play(Write(text_para_left), run_time = 0.5)
            self.play(Write(text_para_right), run_time = 1.5)
            self.wait(10)


# Aligned Animation Part C
        def PartC():
            self.add(ax_3D, para_curve, t0_label, t0_line, t1_label, t1_line, text_para_def_int, t_bar_group, x_label, y_label, z_label, text_para_left, text_para_right)
            self.wait(1)
            self.play(FadeOut(text_para_left), FadeOut(text_para_right), FadeOut(text_para_def_int), run_time = 3)
            self.play(FadeOut(t_bar_group), FadeTransform(x_label, npla_x_group), FadeTransform(y_label, npla_y_group), FadeTransform(z_label, npla_z_group), FadeIn(x_line), FadeIn(y_line), FadeIn(z_line), FadeIn(r_sphere), run_time = 3)
            self.add(x_func_plot, y_func_plot, z_func_plot)
            t_tracker.set_value(t0)
            r_sphere.add_updater(r_updater)
            x_line.add_updater(x_line_updater)
            y_line.add_updater(y_line_updater)
            z_line.add_updater(z_line_updater)
            x_func_plot.add_updater(x_func_updater)
            y_func_plot.add_updater(y_func_updater)
            z_func_plot.add_updater(z_func_updater)
            self.wait(1.5)
            self.play(t_tracker.animate.set_value(t1), rate_func = linear, run_time = 15)
            x_line.remove_updater(x_line_updater)
            y_line.remove_updater(y_line_updater)
            z_line.remove_updater(z_line_updater)
            x_func_plot.remove_updater(x_func_updater)
            y_func_plot.remove_updater(y_func_updater)
            z_func_plot.remove_updater(z_func_updater)
            self.play(FadeOut(r_sphere), FadeOut(x_line), FadeOut(y_line), FadeOut(z_line), run_time = 3)
            self.play(Write(npla_ylabel_x_ext), Write(npla_ylabel_y_ext), Write(npla_ylabel_z_ext), run_time = 1.5)
            self.wait(3)
            self.play(FadeTransform(VGroup(npla_x_group, x_func_plot, npla_ylabel_x_ext, npla_y_group, y_func_plot, npla_ylabel_y_ext, npla_z_group, z_func_plot, npla_ylabel_z_ext), mapping_group_end), run_time = 3)
            self.play(FadeTransform(mapping_group_end, mapping_group_end_white), run_time = 3)
            self.wait(10)

        
        t_tracker = ValueTracker(0)
```

[0:00] Today we are going to parametrize a threedimensional curve.

The curve is the grey spiral on the left side of the screen. We are now trying to describe it mathematically.

[0:14] A scalar variable tea is introduced, running from the spirals start to its end.

While those values can be chosen arbitrarily, they are supposed to make life as easy as possible. We will choose tea to be between zero and ten pi.

---

For each value of tea we now want to receive a specific point on the curve. Namely one individual value for each coordinate.

This relation is a function mapping from a scalar variable to the threedimensional space.

Finding such a function is the essence of parametrization.

---

We will now look at each spatial coordinate individually while letting tee run from teazero to teaone. 

Those functions we already know. Ex of tee is a cosine with amplitude two. Why of tee a plain sine. And zee of tee a basic linear relation.

Going back to the beginning, we can now write down our final parametrization as follows.

```python
%%capture_video --path "animations/parametrization/parametrization_GER_A.mp4"
%%manim -qh --fps 60 $video_scene

class para_Scene(ThreeDScene):
    def construct(self):
        CVC = Text('CVC', font_size = 12, weight = BOLD, color = WHITE, font = 'Latin Modern Sans').align_on_border(RIGHT + DOWN, buff = 0.2)
        self.add(CVC)

         # Headline
        text_parametrization = Tex(r"Parametrisierung einer Trajektorie im $\mathbb{R}^3$", font_size = 48).align_on_border(UP + LEFT, buff = 0.5).shift(0.5 * RIGHT)

        text_para_def = MathTex(r"t\in\left[t_0,\,t_1\right]", font_size = 48).move_to([0.5, 0, 0])
        text_para_def_int = MathTex(r"t\in\left[0,\,10\pi\right]", font_size = 48).move_to([0.5, 0, 0])
        text_para_left = MathTex(r",\quad t\mapsto", font_size = 48).next_to(text_para_def_int, RIGHT)
        text_para_right = Matrix([[r'x(t)'], [r'y(t)'], [r'z(t)']]).scale(0.85).next_to(text_para_left, RIGHT)
        text_para_right_end = Matrix([[r'2\cos{(t)}'], [r'\sin{(t)}'], [r't/(2\pi)']]).scale(0.85).next_to(text_para_left, RIGHT)
        text_para_right_end_white = Matrix([[r'2\cos{(t)}'], [r'\sin{(t)}'], [r't/(2\pi)']]).scale(0.85).next_to(text_para_left, RIGHT)
        text_para_right[0][0].set_color(BLUE)
        text_para_right[0][1].set_color(YELLOW)
        text_para_right[0][2].set_color(RED)
        text_para_right_end[0][0].set_color(BLUE)
        text_para_right_end[0][1].set_color(YELLOW)
        text_para_right_end[0][2].set_color(RED)
        mapping_group = VGroup(text_para_left, text_para_right, text_para_def)
        mapping_group_int = VGroup(text_para_left, text_para_right, text_para_def_int)
        mapping_group_end = VGroup(text_para_left, text_para_right_end, text_para_def_int)
        mapping_group_end_white = VGroup(text_para_left, text_para_right_end_white, text_para_def_int)


        # 3D-Coordinate System
        CO3D = [-6, 0, -8]
        CO3D_x_range = (-3, 3, 1)
        CO3D_y_range = (-3, 3, 1)
        CO3D_z_range = (-1, 6, 1)
        ax = ThreeDAxes(
            x_range = CO3D_x_range, y_range = CO3D_y_range, z_range = CO3D_z_range,
            x_length = 6, y_length = 6, z_length = 9, axis_config = {'tip_length': 0.05, 'tip_width': 0.3}, 
            z_axis_config = {'color': WHITE},
            ).move_to(CO3D).rotate(
                axis = [1, 0, 0], angle = 6*PI/4
                ).rotate(
                    axis = [0, 1, 0], angle = 5*PI/4
                    ).rotate(axis = [1, 0, 0], angle = PI/12).rotate(
                        axis = [0, 1, 0], angle = PI/16
                        )
        x_lab = Tex('$x$', font_size = 64).move_to(CO3D).shift(4*DOWN + 2.75*LEFT)
        y_lab = Tex('$y$', font_size = 64).move_to(CO3D).shift(4*DOWN + 2.75*RIGHT)
        z_lab = Tex('$z$', font_size = 64).move_to(CO3D + 0.5*RIGHT).shift(4.8*UP)
        ax_3D = VGroup(ax, x_lab, y_lab, z_lab).set_opacity(0.5)


        # components of the parametric function
        def x_func(t):
            return 2 * np.cos(t)

        def y_func(t):
            return 1 * np.sin(t)
            
        def z_func(t):
            return t / 2 / PI


        # parametric function specified to ax
        def para_ax_func(t):
            return ax.c2p(x_func(t), y_func(t), z_func(t))

        # generates t_line
        def t_line(t, color):
            return Line3D(para_ax_func(t) + np.array([0, 0.25, 0]), para_ax_func(t) + np.array([0, -0.25, 0]), color = color)

        # generates connector
        def r_connector(start, end, color = RED):
            return Line(start, end, color = color, stroke_width = 4)


        # function start and end
        t0 = 0
        t1 = 10 * PI
        t_bar = 6.75 * PI
        para_curve = ParametricFunction(para_ax_func, color = GREY, t_range = np.array([t0, t1, 0.01])).set_shade_in_3d(True)
        t0_line = t_line(t0, WHITE)
        t1_line = t_line(t1, WHITE)
        t0_label = MathTex(r't_0', font_size = 36, stroke_width = 1, color = WHITE).next_to(t0_line, DOWN)
        t1_label = MathTex(r't_1', font_size = 36, stroke_width = 1, color = WHITE).next_to(t1_line, UP)


        # 3D sphere and connectors
        r_sphere = Sphere(center = para_ax_func(0), radius = 0.1, resolution = (10, 10))
        r_sphere.set_color(GREY)
        #r_sphere.z_index = 5
        x = x_func(0)
        y = y_func(0)
        z = z_func(0)
        x_line = r_connector(ax.c2p(0, y, 0), ax.c2p(x, y, 0), color = BLUE)
        x_line.z_index = 0
        y_line = r_connector(ax.c2p(x, 0, 0), ax.c2p(x, y, 0), color = YELLOW)
        z_line = r_connector(ax.c2p(x, y, z), ax.c2p(x, y, 0), color = RED)
        x_label = MathTex(r"x(t)", font_size = 36, color = BLUE).move_to([-2.75, -2.25, 0])
        y_label = MathTex(r"y(t)", font_size = 36, color = YELLOW).move_to([-3.75, -1.65, 0])
        z_label = MathTex(r"z(t)", font_size = 36, color = RED).move_to([-2.6, -0.5, 0])

        r_sphere_bar = Sphere(center = para_ax_func(t_bar), radius = 0.1, resolution = (10, 10))
        r_sphere_bar.set_color(GREY)
        #r_sphere_bar.z_index = 5
        x_bar = x_func(t_bar)
        y_bar = y_func(t_bar)
        z_bar = z_func(t_bar)
        x_line_bar = r_connector(ax.c2p(0, y_bar, 0), ax.c2p(x_bar, y_bar, 0), color = BLUE)
        x_line_bar.z_index = 0
        y_line_bar = r_connector(ax.c2p(x_bar, 0, 0), ax.c2p(x_bar, y_bar, 0), color = YELLOW)
        z_line_bar = r_connector(ax.c2p(x_bar, y_bar, z_bar), ax.c2p(x_bar, y_bar, 0), color = RED)
        t_bar_group = VGroup(r_sphere_bar, x_line_bar, y_line_bar, z_line_bar)
        t_bar_labels = VGroup(x_label, y_label, z_label)


        # 2D-Coordinate Sytem for each coordinate
        origin_x = [2.5, 2.5, 0]
        origin_y = [2.5, 0, 0]
        origin_z = [2.5, -2.5, 0]
        x_length = 7
        y_length = 1.75
        x_range = [-0.5, 10.5, 1]
        x_dict = dict(zip([5, 10], [r"$5\pi$", r"$10\pi$"]))

        # x(t) system
        y_range_x = [-2.5, 2.5, 1]                          
        y_dict_x = dict(zip([-2, 0, 2], [-2, 0, 2]))
        npla_x = NumberPlane(
            x_range = x_range, y_range = y_range_x, x_length = x_length, y_length = y_length, 
            axis_config = {"include_numbers": False, "stroke_opacity": 0.25, "label_direction": DL}, y_axis_config = {"label_direction": DL}, 
            background_line_style = {"stroke_opacity": 0.25}, ).move_to(origin_x).add_coordinates(x_dict, y_dict_x)#.set_color(BLUE)
        npla_xlabel_x = npla_x.get_x_axis_label(Tex(r"$t$", font_size = 28)).shift(0.1 *LEFT)
        npla_ylabel_x = npla_x.get_y_axis_label(Tex(r"$x(t)$", font_size = 28, color = BLUE))#.shift(1.05 * LEFT + 0.15 * DOWN)
        npla_ylabel_x_ext = Tex(r"$=~2\cos{(t)}$", font_size = 28, color = BLUE).next_to(npla_ylabel_x, RIGHT)
        npla_x_group = VGroup(npla_x, npla_xlabel_x, npla_ylabel_x)

        # y(t) system
        y_range_y = [-1.25, 1.25, 1]                          
        y_dict_y = dict(zip([-1, 0, 1], [-1, 0, 1]))
        npla_y = NumberPlane(
            x_range = x_range, y_range = y_range_y, x_length = x_length, y_length = y_length, 
            axis_config = {"include_numbers": False, "stroke_opacity": 0.25, "label_direction": DL}, y_axis_config = {"label_direction": DL}, 
            background_line_style = {"stroke_opacity": 0.25}, ).move_to(origin_y).add_coordinates(x_dict, y_dict_y)
        npla_xlabel_y = npla_y.get_x_axis_label(Tex(r"$t$", font_size = 28)).shift(0.1 *LEFT)
        npla_ylabel_y = npla_y.get_y_axis_label(Tex(r"$y(t)$", font_size = 28, color = YELLOW))#.shift(1.05 * LEFT + 0.15 * DOWN)
        npla_ylabel_y_ext = Tex(r"$=~\sin{(t)}$", font_size = 28, color = YELLOW).next_to(npla_ylabel_y, RIGHT)
        npla_y_group = VGroup(npla_y, npla_xlabel_y, npla_ylabel_y)

        # z(t) system
        y_range_z = [-0.5, 5.5, 1]                          
        y_dict_z = dict(zip([0, 5], [0, 5]))
        npla_z = NumberPlane(
            x_range = x_range, y_range = y_range_z, x_length = x_length, y_length = y_length, 
            axis_config = {"include_numbers": False, "stroke_opacity": 0.25, "label_direction": DL}, y_axis_config = {"label_direction": DL}, 
            background_line_style = {"stroke_opacity": 0.25}, ).move_to(origin_z).add_coordinates(x_dict, y_dict_z)
        npla_xlabel_z = npla_z.get_x_axis_label(Tex(r"$t$", font_size = 28)).shift(0.1 *LEFT)
        npla_ylabel_z = npla_z.get_y_axis_label(Tex(r"$z(t)$", font_size = 28, color = RED))#.shift(0.9 * LEFT + 0.15 * DOWN)
        npla_ylabel_z_ext = Tex(r"$=~\frac{t}{2\pi}$", font_size = 28, color = RED).next_to(npla_ylabel_z, RIGHT)
        npla_z_group = VGroup(npla_z, npla_xlabel_z, npla_ylabel_z)


        # 2D-functions plus their updaters
        x_func_plot = npla_x.plot(lambda t_var: x_func(t_var * PI), color = BLUE, stroke_width = 1.5, stroke_opacity = 1, x_range = [0, 0])
        y_func_plot = npla_y.plot(lambda t_var: y_func(t_var * PI), color = YELLOW, stroke_width = 1.5, stroke_opacity = 1, x_range = [0, 0])
        z_func_plot = npla_z.plot(lambda t_var: z_func(t_var * PI), color = RED, stroke_width = 1.5, stroke_opacity = 1, x_range = [0, 0])

        def x_func_updater(func):
            t = t_tracker.get_value()
            func.become(npla_x.plot(lambda t_var: x_func(t_var * PI), color = BLUE, stroke_width = 1.5, stroke_opacity = 1, x_range = [0, 0.0001 + t / PI]))

        def y_func_updater(func):
            t = t_tracker.get_value()
            func.become(npla_y.plot(lambda t_var: y_func(t_var * PI), color = YELLOW, stroke_width = 1.5, stroke_opacity = 1, x_range = [0, 0.0001 + t / PI]))

        def z_func_updater(func):
            t = t_tracker.get_value()
            func.become(npla_z.plot(lambda t_var: z_func(t_var * PI), color = RED, stroke_width = 1.5, stroke_opacity = 1, x_range = [0, 0.0001 + t / PI]))


        # updates sphere and connectors
        def r_updater(sphere):
            t = t_tracker.get_value()
            x = x_func(t)
            y = y_func(t)
            z = z_func(t)
            sphere.move_to(para_ax_func(t))
       
        def x_line_updater(line):
            t = t_tracker.get_value()
            x = x_func(t)
            y = y_func(t)
            z = z_func(t)
            line.become(r_connector(ax.c2p(0, y, 0), ax.c2p(x, y, 0), color = BLUE))

        def y_line_updater(line):
            t = t_tracker.get_value()
            x = x_func(t)
            y = y_func(t)
            z = z_func(t)
            line.become(r_connector(ax.c2p(x, 0, 0), ax.c2p(x, y, 0), color = YELLOW))

        def z_line_updater(line):
            t = t_tracker.get_value()
            x = x_func(t)
            y = y_func(t)
            z = z_func(t)
            line.become(r_connector(ax.c2p(x, y, z), ax.c2p(x, y, 0), color = RED))


# Aligned Animation Part A
        def PartA():
            self.wait(1)
            self.play(Write(text_parametrization), run_time = 2)
            self.wait(1.5)
            self.play(FadeOut(text_parametrization), FadeIn(ax_3D), run_time = 3)
            self.play(Create(para_curve), run_time = 4)
            self.wait(4.5)
            self.play(Write(text_para_def), run_time = 1.5)
            self.wait(2.5)
            self.play(Write(t0_label), FadeIn(t0_line), run_time = 1.5)
            self.play(Write(t1_label), FadeIn(t1_line), run_time = 1.5)
            self.wait(9)
            self.play(FadeTransform(text_para_def, text_para_def_int), run_time = 3)
            self.wait(1.5)


# Aligned Animation Part B
        def PartB():
            self.add(ax_3D, para_curve, t0_label, t0_line, t1_label, t1_line, text_para_def_int)
            self.wait(2)
            self.play(FadeIn(t_bar_group), run_time = 5)
            self.play(FadeIn(t_bar_labels), run_time = 3)
            self.wait(4)
            self.play(Write(text_para_left), run_time = 0.5)
            self.play(Write(text_para_right), run_time = 1.5)
            self.wait(10)


# Aligned Animation Part C
        def PartC():
            self.add(ax_3D, para_curve, t0_label, t0_line, t1_label, t1_line, text_para_def_int, t_bar_group, x_label, y_label, z_label, text_para_left, text_para_right)
            self.wait(1)
            self.play(FadeOut(text_para_left), FadeOut(text_para_right), FadeOut(text_para_def_int), run_time = 3)
            self.play(FadeOut(t_bar_group), FadeTransform(x_label, npla_x_group), FadeTransform(y_label, npla_y_group), FadeTransform(z_label, npla_z_group), FadeIn(x_line), FadeIn(y_line), FadeIn(z_line), FadeIn(r_sphere), run_time = 3)
            self.add(x_func_plot, y_func_plot, z_func_plot)
            t_tracker.set_value(t0)
            r_sphere.add_updater(r_updater)
            x_line.add_updater(x_line_updater)
            y_line.add_updater(y_line_updater)
            z_line.add_updater(z_line_updater)
            x_func_plot.add_updater(x_func_updater)
            y_func_plot.add_updater(y_func_updater)
            z_func_plot.add_updater(z_func_updater)
            self.wait(1.5)
            self.play(t_tracker.animate.set_value(t1), rate_func = linear, run_time = 15)
            x_line.remove_updater(x_line_updater)
            y_line.remove_updater(y_line_updater)
            z_line.remove_updater(z_line_updater)
            x_func_plot.remove_updater(x_func_updater)
            y_func_plot.remove_updater(y_func_updater)
            z_func_plot.remove_updater(z_func_updater)
            self.play(FadeOut(r_sphere), FadeOut(x_line), FadeOut(y_line), FadeOut(z_line), run_time = 3)
            self.play(Write(npla_ylabel_x_ext), Write(npla_ylabel_y_ext), Write(npla_ylabel_z_ext), run_time = 1.5)
            self.wait(3)
            self.play(FadeTransform(VGroup(npla_x_group, x_func_plot, npla_ylabel_x_ext, npla_y_group, y_func_plot, npla_ylabel_y_ext, npla_z_group, z_func_plot, npla_ylabel_z_ext), mapping_group_end), run_time = 3)
            self.play(FadeTransform(mapping_group_end, mapping_group_end_white), run_time = 3)
            self.wait(10)

        
        t_tracker = ValueTracker(0)
        PartA()
```
